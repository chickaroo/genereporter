```python
# look at distribution
def plot_hist(df):
    hist, edges = np.histogram(df, bins=int(np.max(df.values)))

    plt.figure()

    plt.bar(edges[:-1], hist, width = 0.5, color='#0504aa')
    plt.xlim(min(edges), 50)
    #plt.ylim(0, len(df)*2)
    plt.grid(axis='y', alpha=0.75)
    plt.xlabel('Gene Expression',fontsize=15)
    plt.ylabel('Frequency',fontsize=15)
    plt.xticks(fontsize=15)
    plt.yticks(fontsize=15)
    plt.show()
```

```python
plot_hist(adata.to_df(layer='tc_norm'))
```
